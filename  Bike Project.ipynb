```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
```

```python
show=pd.read_csv("E:\\Data Projects\\BIKE DETAILS.csv")
pd.set_option('display.max_rows', None)
show
```

```python
 show.isnull().sum()
```

```python
show["ex_showroom_price"].mean()
```

```python
show=show.fillna(show["ex_showroom_price"].mean())     #mean fill in it
```

```python
show.isnull().sum()         # here see clean and fill all null values
```

```python
show["selling_price"].mean()         #also s\check with this
```

```python
show["Current P or L"]=show["selling_price"] - show["ex_showroom_price"]
```

```python
show
```

# Year higgest Number of Sales

```python
show.groupby("year").sum().sort_values(["selling_price"],ascending=False)
```

```python
# now with plot
plt.figure(figsize=(10,7))
sns.lineplot(x="year",y="selling_price",data=show,hue="seller_type")
plt.show()
```

# Only Royal Company Bikes


```python
disp= show.loc[show['name'].str.contains("Royal", case=False)]
```

```python
disp
```

```python
print(len(disp))          # 101 royal Bikes in the dataset
```

```python
 disp.groupby("name")["selling_price","km_driven"].sum().sort_values(["selling_price"],ascending=False)
```

```python
#disp['final'] = disp.groupby('name')['selling_price'].transform('mean')      # see transform and use aggrate function
#disp
```

```python
plt.figure(figsize=(10,6))
bins=[50000,100000,150000,200000,250000,300000]
sns.distplot(disp["selling_price"],bins=bins,color="r",rug=True)
plt.title("Royal Bikes Price Analysis",fontsize=20)
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x="year",y="selling_price",data=disp,hue="owner")
plt.show()
```

```python
disp.describe()
```

# now filtering

```python
filtr=disp.loc[(disp["owner"]=="2nd owner") & (disp["km_driven"]>=40000)]
filtr
```

```python
sns.distplot(filtr["km_driven"])
```

```python
plt.figure(figsize=(16,9))
order=["1st owner","2nd owner","3rd owner"]
more=sns.barplot(x="name",y="selling_price",data=disp,estimator=np.max,hue="owner",hue_order=order)
plt.xticks(rotation=90)
more.set(title="Royal Bike Analysis")
plt.show()
```
